```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
passengers = pd.read_csv('data/passengers.csv', index_col='date', parse_dates=True)
```

```python
passengers['num_passengers'].plot(figsize=(12,5))
plt.title('International airline passengers: monthly totals in thousands. Jan 49 – Dec 60')
plt.show()
```

```python
passengers.head(5)
```

```python
passengers.tail(5)
```

```python
month = np.arange(12)
```

```python
new_col=[]
i=1949
while i<=1960:
    new_col=np.concatenate([new_col,month])
    i=i+1
new_col
```

```python
one = np.ones(144,dtype=int)
len(one)
```

```python
passengers['one']=one
```

```python
passengers['month']=new_col
```

```python
def sin_month(i):
    return np.sin((2*np.pi *i)/12)
```

```python
def cos_month(i):
    return np.cos((2*np.pi *i)/12)
```

```python
a=[]
for j in month:
    a.append(sin_month(j))
a
new_col_sin=[]
i=1949
while i<=1960:
    new_col_sin=np.concatenate([new_col_sin,a])
    i=i+1
```

```python
b=[]
for j in month:
    b.append(cos_month(j))
b
new_col_cos=[]
i=1949
while i<=1960:
    new_col_cos =np.concatenate([new_col_cos,b])
    i=i+1
```

```python
num=np.arange(0,144,dtype=int)
```

```python
passengers['Sin']=new_col_sin
passengers['Cos']=new_col_cos
passengers['Num']=num
```

```python
passengers
```

```python
pas_val=passengers[['one','Num','Sin','Cos']].values
pas_val
```

```python
from scipy.optimize import minimize
```

```python
def MSE(Q1, Q2, b1, b2):
    res = np.dot(pas_val, np.array([[Q1],[Q2],[b1], [b2]]))
    t = len(pas_val)
    return sum((res.reshape(t) - passengers["num_passengers"].values)**2)/t
```

```python
# МНК
result = minimize(lambda x: MSE(x[0], x[1], x[2], x[3]), (0.5, 0.5, 0.5, 0.5))
```

```python
Q1, Q2, b1, b2 = result.x
Q1, Q2, b1, b2
```

```python
predictions = np.dot(pas_val, np.array([[Q1],[Q2],[b1], [b2]])).reshape(len(pas_val))
predictions
```

```python
plt.plot(predictions)
plt.plot(passengers["num_passengers"].values)
plt.show()
```
